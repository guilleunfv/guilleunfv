```python
import pandas as pd
import matplotlib.pyplot as plt

# Ler o arquivo CSV e carregar os dados em um DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/balaiocientifico/dataset/main/healthexp.csv')

# Calcular a média de gastos por país
avg_spending = df.groupby('Country')['Spending_USD'].mean().reset_index()

# Ordenar os países por média de gastos em ordem decrescente
avg_spending = avg_spending.sort_values('Spending_USD', ascending=False)

# Criar o gráfico de barras
plt.bar(avg_spending['Country'], avg_spending['Spending_USD'])
plt.xlabel('País')
plt.ylabel('Gasto médio em saúde (USD)')
plt.title('Gasto médio em saúde por país')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()





```

Gráfico de barras (Gráfico de barras) para mostrar a média de gastos com saúde por país:
O gráfico de barras apresenta a média de gastos com saúde em cada país. Cada barra representa um país e a altura da barra indica o valor médio de gastos com saúde nesse país. As barras estão organizadas em ordem decrescente com base nos gastos médios, permitindo comparar visualmente os países que gastam mais e menos em saúde.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ler o arquivo CSV e carregar os dados em um DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/balaiocientifico/dataset/main/healthexp.csv')

# Calcular o total de gastos por ano
total_spending = df.groupby('Year')['Spending_USD'].sum()

# Criar o gráfico de setores
plt.pie(total_spending, labels=total_spending.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Proporção de gastos com saúde por ano')
plt.tight_layout()
plt.show()

```

Gráfico de setores (Gráfico de pizza) para mostrar a proporção de gastos com saúde por ano:
O gráfico de setores exibe a proporção de gastos com saúde em cada ano. Cada fatia do gráfico de pizza representa um ano e a área da fatia corresponde à proporção do total de gastos com saúde naquele ano. A legenda mostra o ano correspondente a cada fatia e a porcentagem correspondente a essa proporção, permitindo uma compreensão visual rápida da distribuição dos gastos com saúde ao longo do tempo.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ler o arquivo CSV e carregar os dados em um DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/balaiocientifico/dataset/main/healthexp.csv')

# Criar o gráfico de linhas
plt.plot(df['Year'], df['Life_Expectancy'])
plt.xlabel('Ano')
plt.ylabel('Expectativa de vida')
plt.title('Evolução da expectativa de vida ao longo dos anos')
plt.grid(True)
plt.tight_layout()
plt.show()


```

Gráfico de linhas (Gráfico de linhas) para mostrar a evolução da expectativa de vida ao longo dos anos:
O gráfico de linhas ilustra a evolução da expectativa de vida ao longo dos anos. Cada ponto no gráfico representa um ano, e a posição vertical do ponto indica o valor da expectativa de vida naquele ano. Traçando uma linha conectando os pontos, é possível observar a tendência geral de mudança na expectativa de vida ao longo do tempo. O gráfico de linhas facilita a identificação de períodos de aumento ou diminuição da expectativa de vida e permite fazer comparações entre diferentes anos.
